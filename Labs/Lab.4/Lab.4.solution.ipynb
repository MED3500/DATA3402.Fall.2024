```python
import math
import random

class shape:
    __x = int()
    __y = int()
    
    def __init__(self, x, y):
        self.__x = x
        self.__y = y

    def random_points_on_perimeter(self, num_points):
        if num_points > 16: return print("too many points, please input a number no bigger than 16")
        list_of_points = []
        while len(list_of_points) < num_points:
            randX = random.randint(self.__x, self.__x)
            randY = random.randint(self.__y, self.__y)
            if (randX <= self.__x) & (randY <= self.__y):
                if (randX == self.__x) | (randY == self.__y):
                    list_of_points.append([randX, randY])
        return list_of_points

    def test_if_inside_shape(self, x, y):
        if (x == self.__x) & (y == self.__y):
            return True
        return False

    def overlapping(self, other):
        testlist = self.random_points_on_perimeter(16)
        is_overlapping = 0
        for i in testlist:
            is_overlapping += (other.test_if_inside_shape(i[0], i[1]))
        return bool(is_overlapping)

class rectangle(shape):
    __length = int()
    __width = int()
    __x = int()
    __y = int()

    def __init__(self, length, width, x, y):
        self.__length = length
        self.__width = width
        self.__x = x
        self.__y = y
        

    def area(self): return self.__length * self.__width
        
    def perimeter(self): return (self.__length *2) + (self.__width *2)

    def get_x_coor(self): return self.__x

    def get_y_coor(self): return self.__y

    def get_length(self): return self.__length

    def get_width(self): return self.__width

    def random_points_on_perimeter(self, num_points):
        if num_points > 16: return "too many points, please input a number no bigger than 16"
        list_of_points = []
        x = self.__x
        y = self.__y
        width = self.__width
        length = self.__length
        
        while len(list_of_points) < num_points:
            rand_x = random.randint(x, x + width)
            rand_y = random.randint(y, y + length)
            if (rand_x == x) | (rand_y == y) | (rand_x == x + width) | (rand_y == y + length):
                list_of_points.append([rand_x, rand_y])
        return list_of_points

    def test_if_inside_shape(self, x, y):
        if (x >= self.__x) & (x <= self.__width) & (y >= self.__y) & (y <= self.__length):
            return True
        return False
        

class circle(shape):
    __radius = int()
    __x = int()
    __y = int()

    def __init__(self, radius, x, y):
        self.__radius = radius
        self.__x = x
        self.__y = y
        

    def area(self): return math.pi * (self.__radius **2)
        
    def perimeter(self): return 2 * math.pi * self.__radius

    def get_x_coor(self): return self.__x

    def get_y_coor(self): return self.__y

    def get_radius(self): return self.__radius

    def random_points_on_perimeter(self, num_points):
        if num_points > 16: return "too many points, please input a number no bigger than 16"
        r = self.__radius
        h = self.__x
        k = self.__y
        list_of_points = []
        while len(list_of_points) < num_points:
            rand_x = random.randint(self.__x - self.__radius, self.__x + self.__radius)
            while rand_x == 0:
                rand_x = random.randint(self.__x - self.__radius, self.__x + self.__radius)
            # (x – h)2 + (y – k)2 = r2
            # y = sqrt((r**2) / ((x-h)**2))+k
            calced_y = math.sqrt((r**2) - ((rand_x-h)**2))+k

            list_of_points.append([rand_x, calced_y])
        return list_of_points

    def test_if_inside_shape(self, x, y):
        r = self.__radius
        h = self.__x
        k = self.__y
            # (x – h)2 + (y – k)2 = r2
            # y = sqrt((r**2) - ((x-h)**2))+k
            # x = sqrt((r**2) - ((y-k)**2)) + h
        if (abs(y) <= math.sqrt((r**2) - ((x-h)**2))+k) & (abs(x) <= math.sqrt((r**2) - ((y-k)**2)) + h):
            return True
        return False


class triangle(shape):
    __side1 = int()
    __side2 = int()
    __side3 = int()
    __x1 = int()
    __y1 = int()
    __x2 = int()
    __y2 = int()
    __x3 = int()
    __y3 = int()
    
    def __init__(self, side1, side2, side3, x1, y1, x2, y2, x3, y3):
        self.__side1 = side1
        self.__side2 = side2
        self.__side3 = side3
        self.__x1 = x1
        self.__y1 = y1
        self.__x2 = x2
        self.__y2 = y2
        self.__x3 = x3
        self.__y3 = y3
        

    def area(self, x1, y1, x2, y2, x3, y3):
        return abs((x1 * (y2 - y3) + x2 * (y3 - y1) + x3 * (y1 - y2)) / 2.0)
        
    def perimeter(self): return self.__side1 + self.__side2 + self.__side3

    def get_coors(self): return [[self.__x1, self.__y1], [self.__x2, self.__y2], [self.__x3, self.__y3]]

    def get_sides(self): return [self.__side1, self.__side2, self.__side3]

    def random_points_on_perimeter(self, num_points):
        if num_points > 16: return "too many points, please input a number no bigger than 16"
            
        x1 = self.__x1
        y1 = self.__y1
        x2 = self.__x2
        y2 = self.__y2
        x3 = self.__x3
        y3 = self.__y3
        
        list_of_points = []

        while len(list_of_points) < num_points:
            rand_side = random.randint(1, 3)

            if rand_side == 1:
                rand_x = random.uniform(x1, x2)
                if x1 == x2:  # Vertical line
                    rand_y = random.uniform(min(y1, y2), max(y1, y2))
                else:
                    slope = (y2 - y1) / (x2 - x1)
                    intercept = y1 - slope * x1
                    rand_y = slope * rand_x + intercept

            elif rand_side == 2:
                rand_x = random.uniform(x2, x3)
                if x2 == x3:  # Vertical line
                    rand_y = random.uniform(min(y2, y3), max(y2, y3))
                else:
                    slope = (y3 - y2) / (x3 - x2)
                    intercept = y2 - slope * x2
                    rand_y = slope * rand_x + intercept

            else:
                rand_x = random.uniform(x3, x1)
                if x3 == x1:  # Vertical line
                    rand_y = random.uniform(min(y3, y1), max(y3, y1))
                else:
                    slope = (y1 - y3) / (x1 - x3)
                    intercept = y3 - slope * x3
                    rand_y = slope * rand_x + intercept

            list_of_points.append([rand_x, rand_y])

        return list_of_points

    def test_if_inside_shape(self, px, py):
        x1 = self.__x1
        y1 = self.__y1
        x2 = self.__x2
        y2 = self.__y2
        x3 = self.__x3
        y3 = self.__y3
            
        A = self.area(x1, y1, x2, y2, x3, y3)

        # Calculate the area of the triangles formed with the point
        A1 = self.area(px, py, x2, y2, x3, y3)
        A2 = self.area(x1, y1, px, py, x3, y3)
        A3 = self.area(x1, y1, x2, y2, px, py)

        # Check if the sum of the areas of A1, A2, and A3 equals A
        return A == A1 + A2 + A3
        



# Testing:

circ = circle(2, 0, 0)
rect = rectangle(1,1,0,0)
tri = triangle(1,1,1,    0,0,     1,0,      0,1)

print("Circle / Rectangle:\t", circ.overlapping(rect))  # False
print("Circle / Triangle:\t", circ.overlapping(tri))    # False
print("Rectangle / Triangle:\t", rect.overlapping(tri)) # False
```

```python
class foo:
    def __init__(self,a,b=None):
        self.a=a
        self.b=b
        
    def __repr__(self):
        return "foo("+repr(self.a)+","+repr(self.b)+")"
    
    def save(self,filename):
        f=open(filename,"w")
        f.write(self.__repr__())
        f.close()
        
   
def foo_loader(filename):
    f=open(filename,"r")
    tmp=eval(f.read())
    f.close()
    return tmp
```

```python
import random

class Deck():
    deck = []
    num_decks = 1
    
    def __init__(self, number_of_decks):
        self.num_decks = number_of_decks
        
        for i in range(number_of_decks):
            for suit in range(1, 5):
                for card in range(1, 14): 
                    if   suit == 1: suit = "Hearts"
                    elif suit == 2: suit = "Clubs"
                    elif suit == 3: suit = "Diamonds"
                    elif suit == 4: suit = "Spades"
        
                    if   card == 1:  card = "Ace"
                    elif card == 11: card = "Jack"
                    elif card == 12: card = "Queen"
                    elif card == 13: card = "King"
        
                    self.deck.append((suit, card))
        
            self.deck.append(("Card","Plastic"))

    def shuffle(self):
        shuffled = []

        for i in range(len(self.deck)):
            rand = random.randint(0,len(self.deck)-1)
            shuffled.append(self.deck[rand])
            self.deck.remove(self.deck[rand])

        for i in shuffled:
            self.deck.append(i)

        for i in self.deck:
            if self.deck.count(i) > self.num_decks:
                print("WARNING: DUPLICATED VALUE (", i, ")!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!")

    def draw(self):
        card = self.deck[0]
        self.deck.remove(card)
        return card


deck1 = Deck(2)
for i in deck1.deck: print(i)

print("\nshuffled\n")

deck1.shuffle()
for i in deck1.deck: print(i)

print("\ndraw first card\n")

print(deck1.draw())

print("\ndeck with card removed\n")

for i in deck1.deck: print(i)

## Shuffle when plastic card will be implimented in later sections for code readability
```

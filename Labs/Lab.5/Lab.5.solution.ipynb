3. Write functions that create special matrices (note these are standalone functions, not member functions of your `matrix` class):
    * `constant(n,m,c)`: returns a `n` by `m` matrix filled with floats of value `c`.
    * `zeros(n,m)` and `ones(n,m)`: return `n` by `m` matrices filled with floats of value `0` and `1`, respectively.
    * `eye(n)`: returns the n by n identity matrix.

```python
def constant(n, m, c):
    matrix = []
    for i in range(n):
        matrix.append([])
        for k in range(m):
            matrix[i].append(c)
    return matrix

def zeros(n, m):
    matrix = []
    for i in range(n):
        matrix.append([])
        for k in range(m):
            matrix[i].append(0)
    return matrix

def ones(n,m):
    matrix = []
    for i in range(n):
        matrix.append([])
        for k in range(m):
            matrix[i].append(1)
    return matrix

def eye(n):
    matrix = []
    for i in range(n):
        matrix.append([])
        for k in range(n):
            matrix[i].append(0)
        matrix[i][i] = 1
    return matrix

def printmatrix(matrix):
    for i in matrix: print(i)

printmatrix(constant(2,3,5))
print()
printmatrix(zeros(2,3))
print()
printmatrix(ones(1,3))
print()
printmatrix(eye(4))
```
